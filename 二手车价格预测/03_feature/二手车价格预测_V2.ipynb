```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

## 加载数据

```python
origin_train = pd.read_csv('../01_data/used_car_train_20200313.csv', sep=' ')
origin_test = pd.read_csv('../01_data/used_car_testB_20200421.csv', sep=' ')
```

```python
result = pd.DataFrame()
result['SaleID'] = origin_test['SaleID']
```

```python
origin_train = origin_train[['regDate', 'creatDate']]
origin_test = origin_test[['regDate', 'creatDate']]
```

```python
train_data = None
with open('../02_user_data/train_v1.pk', mode='rb') as f:
    train_data = pickle.load(f)
train_data
```

```python
test_data = None
with open('../02_user_data/test_v1.pk', mode='rb') as f:
    test_data = pickle.load(f)
test_data
```

## 缺失值填充

```python
temp = train_data.isnull().sum()
temp = temp[temp > 0]
temp
```

***这里用众数填充***

```python
for feature in list(temp.index):
    train_data[feature].fillna(train_data[feature].mode()[0], inplace=True)
    test_data[feature].fillna(test_data[feature].mode()[0], inplace=True)
```

```python
train_data.isnull().sum()
```

```python
test_data.isnull().sum()
```

## 特征处理

#### 时间相关（regDate, creatDate）

时间多尺度处理

```python
origin_train['regDate'] = pd.to_datetime(origin_train['regDate'])
train_data['regDate_Y'] = origin_train['regDate'].dt.year
train_data['regDate_M'] = origin_train['regDate'].dt.month
train_data['regDate_D'] = origin_train['regDate'].dt.day
train_data['regDate_Diff'] = (origin_train['regDate'] - np.min(origin_train['regDate'])).dt.days
```

```python
train_data.info()
```

```python
origin_test.info()
```

```python
origin_test['regDate'] = pd.to_datetime(origin_test['regDate'])
test_data['regDate_Y'] = origin_test['regDate'].dt.year
test_data['regDate_M'] = origin_test['regDate'].dt.month
test_data['regDate_D'] = origin_test['regDate'].dt.day
test_data['regDate_Diff'] = (origin_test['regDate'] - np.min(origin_test['regDate'])).dt.days
```

```python
origin_train['creatDate'] = pd.to_datetime(origin_train['creatDate'])
train_data['creatDate_Y'] = origin_train['creatDate'].dt.year
train_data['creatDate_M'] = origin_train['creatDate'].dt.month
train_data['creatDate_D'] = origin_train['creatDate'].dt.day
train_data['creatDate_Diff'] = (origin_train['creatDate'] - np.min(origin_train['creatDate'])).dt.days
```

```python
origin_test['creatDate'] = pd.to_datetime(origin_test['creatDate'])
test_data['creatDate_Y'] = origin_test['creatDate'].dt.year
test_data['creatDate_M'] = origin_test['creatDate'].dt.month
test_data['creatDate_D'] = origin_test['creatDate'].dt.day
test_data['creatDate_Diff'] = (origin_test['creatDate'] - np.min(origin_test['creatDate'])).dt.days
```

```python
test_data.isnull().sum()
```

```python
# 模型创建 - XGBoost
import xgboost as xgb
param = {
    #'objective' : 'reg:linear',
    'eval_metric' : 'auc',
    'tree_method' : 'gpu_hist',
    'max_depth' : 6,
    'colsample_bytree': 0.8,
    'min_child_samples' : 3,
    'subsample' : 0.8,
    'reg_lambda' : 0.5,
    'learning_rate' : 0.05,
    'n_estimators' : 2000,
}
model = xgb.XGBRegressor(
    max_depth=6, learning_rate=0.05, n_estimators=300, 
    objective='reg:linear', tree_method='gpu_hist', 
    subsample=0.8, colsample_bytree=0.8, 
    min_child_samples=3, eval_metric='auc', reg_lambda=0.5
)
```

```python
X_train = train_data.drop(columns=['price'])
Y_test = train_data['price']
X_test = test_data
```

```python
model.fit(X_train, Y_test)
```

```python
y_predictions = model.predict(X_test)
y_predictions
```

```python
def show_status(data):
    print('max: {}'.format(np.max(data)))
    print('min: {}'.format(np.min(data)))
    print('ptp: {}'.format(np.ptp(data)))
    print('std: {}'.format(np.std(data)))
    print('mean: {}'.format(np.mean(data)))
    print('var: {}'.format(np.var(data)))
```

```python
show_status(y_predictions)
```

```python
result['price'] = y_predictions
temp = result['price']
temp[temp < 0] = 11
result['price'] = temp
show_status(result['price'])
```

```python
result.to_csv('../05_prediction_result/xgb_predictions_v2.csv')
```
